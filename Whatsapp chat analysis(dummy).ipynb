```python
import re
import pandas as pd
```

```python
f = open("WhatsApp Chat with CC Group Ms. Poonamjot[142].txt" , 'r' , encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern , data)
dates
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})
```

```python
df.head(3)
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M - ')
```

```python
df.shape
```

```python
df.rename(columns = {'message_date':'date'}, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.head()
```

```python
df['year']  =df['date'].dt.year
```

```python
temp = df[df['user']!='group_notification']
temp.head()
```

```python
temp = temp[temp['message'] !='<Media omitted>\n']
```

```python
temp.head()
```

```python
f = open('Hinglish_stop_words.txt','r')
stop_words = f.read()
print(stop_words)
```

```python
# how to count the words
words = []
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
words
```

```python
from collections import Counter
Counter(words)
```

```python
#how to find most common 20 words
Counter(words).most_common(20)
```

```python
Most_common_words = pd.DataFrame(Counter(words).most_common(20))
Most_common_words
```

```python
Most_common_words.rename(columns = {0:'Most Common Words', 1:'Count'}, inplace = True)
```
